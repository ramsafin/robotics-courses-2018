```matlab
% 1.1. Using a 3D transformation matrix translate point (2,6,0)
% by 4 in the x-direction and by -3 in the y-direction.
% Write down the corresponding transformation homogeneous matrix.

P = [2; 6; 0]; % coordinates of the vector from the origin to point (2,6,0)

tx = 4;
ty = -3;

T = [tx; ty; 0]; % 3D translation vector

H = transl(tx, ty, 0) % correspondant homogeneous matrix

Q1 = P + T;
Q2 = H * [P; 1]; % translated vector coordinates P + T ~ H * (2; 6; 0; 1)

% draw the initial point and the point after transformation

axis([0 7 0 7 0 1])
hold on
grid on

plot3(P(1), P(2), P(3), '*', 'color', 'b')
plot3(Q1(1), Q1(2), Q1(3), '*', 'color', 'g')
plot3(Q2(1), Q2(2), Q2(3), 'o', 'color', 'r')

text(P(1) - 0.8, P(2) - 0.5, P(3), 'P (2,6,0)')
text(Q1(1) - 0.8, Q1(2) - 0.5, Q1(3), 'Q1')
text(Q2(1), Q2(2) + 1, Q2(3), 'Q2')

axis square
xlabel('x'), ylabel('y'), zlabel('z')
```

```matlab
% 1.2. Using a 3D transformation matrix rotate point (2,6,0)
% by 90 degrees anti-clockwise about (0,0), i.e. Z-axis.
% Write down the corresponding transformation homogeneous matrix.

P = [2; 6; 0]; % coordinates of a vector to a point (2,6,0)

R = rotz(90); % 90 degrees rotation around Z-axis matrix

H = rotm2tform(R) % correspondant homogeneous matrix

Q1 = R * P; % rotated vector coordinates

Q2 = H * [P; 1]; % same as above, i.e. Q1 ~ Q2

axis([-7 7 0 7 0 1])
hold on
grid on

plot3(P(1), P(2), P(3), '*')
text(P(1) + 1, P(2), P(3), 'P (2,6,0)')

plot3(Q1(1), Q1(2), Q1(3), '*')
text(Q1(1) + 1, Q1(2) + 1, Q1(3), 'Q1 (-6,2,0)')

plot3(Q2(1), Q2(2), Q2(3), 'o', 'color', 'b')
text(Q2(1), Q2(2) - 0.5, Q2(3), 'Q2')

grid on
axis square
xlabel('x'), ylabel('y'), zlabel('z')
```
